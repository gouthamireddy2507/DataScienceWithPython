```python
import numpy as np
import matplotlib.pyplot as ply
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
a
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y')
axes[1][0].set_title('Plot3')
axes[1][1].plot(a,b)
fig.suptitle("FIGURE TITLE",fontsize=16)
fig.subplots_adjust(wspace=0.3,hspace=0.3)
# plt.tight_layout()
fig.savefig('new_subplots.png')
```
